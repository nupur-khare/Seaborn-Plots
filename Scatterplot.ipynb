```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
diamonds = diamonds[diamonds.cut.isin(['Premium', 'Good']) & diamonds.color.isin(['D','F', 'J'])].sample(n=100, random_state=22)
```

```python
sns.set_style('dark')
```

```python
sns.scatterplot(x=diamonds.carat, y=diamonds.price)
```

```python
sns.set_style('darkgrid')
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds, hue='depth', s=60)
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds, style='cut', s=100)
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds, hue='depth', size='depth')
```

```python
sns.scatterplot(x='carat', y='price', hue='cut', style='color', data=diamonds, s=100)
```

```python
sns.scatterplot(x=diamonds.carat, y=diamonds.price, alpha=0.8, s=100)
```

```python
 sns.scatterplot(x=diamonds.carat, y=diamonds.price, s=200, marker='*', edgecolor='blue')
```

```python
sns.lineplot(x='depth', y='table', data=diamonds)
```
