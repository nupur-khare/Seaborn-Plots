```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
diamonds = diamonds[diamonds.clarity.isin(['SI1','VS1'])]
```

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(y='color', data=diamonds)
```

```python
sns.countplot(x='color', data=diamonds, hue='clarity')
```

```python
sns.countplot(x='color', data=diamonds, hue='clarity', hue_order=['SI1', 'VS1'])
```

```python
sns.countplot(x='color', data=diamonds, color='purple')
```

```python
sns.countplot(x='color', data=diamonds, palette='twilight')
```

```python
sns.countplot(x='color', data=diamonds, palette='twilight', lw=4, ec='black',hatch='*')
```
