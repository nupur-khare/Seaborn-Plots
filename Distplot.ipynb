```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('white')
```

```python
sns.displot(cars.displacement, kde=False, rug=True, bins=20)
```

```python
sns.displot(cars.displacement, kde=False, rug=True, bins=[160, 180, 200, 360])
```

```python
sns.displot(cars.weight, kde_kws={'lw':5});
```

```python
sns.displot(cars.weight, kde_kws={'bw_method':10}, rug=True, rug_kws={'height':0.15, 'color':'r'});
```
