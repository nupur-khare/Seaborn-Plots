```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.head()
```

```python
sns.set_style('darkgrid')
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement)
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement, data=cars, hue='origin');
plt.legend(loc=2);
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement, data=cars, split=True, inner='quartiles', scale='count')
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement, bw=0.2)
```
